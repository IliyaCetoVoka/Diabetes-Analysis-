```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("diabetes.csv")

# 1. Histogram
plt.figure(figsize=(6,4))
plt.hist(df["Glucose"], bins=20)
plt.title("Distribution of Glucose Levels")
plt.xlabel("Glucose Level")
plt.ylabel("Frequency")
plt.show()

# 2. Scatter Plot
plt.figure(figsize=(6,4))
plt.scatter(df["Age"], df["Glucose"])
plt.title("Age vs Glucose Level")
plt.xlabel("Age")
plt.ylabel("Glucose Level")
plt.show()

# 3. Bar Chart
outcome_counts = df["Outcome"].value_counts()
plt.figure(figsize=(6,4))
plt.bar(outcome_counts.index, outcome_counts.values)
plt.title("Diabetes Outcome Count")
plt.xlabel("Outcome")
plt.ylabel("Number of Patients")
plt.show()

# 4. Box Plot
plt.figure(figsize=(6,4))
plt.boxplot(df["BMI"].dropna())
plt.title("BMI Distribution")
plt.xlabel("BMI")
plt.ylabel("BMI Value")
plt.show()

# 5. Line Chart
plt.figure(figsize=(6,4))
plt.plot(df["Glucose"].head(50))
plt.title("Glucose Levels of First 50 Patients")
plt.xlabel("Patient Index")
plt.ylabel("Glucose Level")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("diabetes.csv")

# Apply seaborn styling
sns.set_style("darkgrid")

# 1. Histogram
plt.figure(figsize=(6,4))
sns.histplot(df["Glucose"], bins=20)
plt.title("Distribution of Glucose Levels")
plt.xlabel("Glucose Level")
plt.ylabel("Frequency")
plt.show()

# 2. Scatter Plot
plt.figure(figsize=(6,4))
sns.scatterplot(x=df["Age"], y=df["Glucose"])
plt.title("Age vs Glucose Level")
plt.xlabel("Age")
plt.ylabel("Glucose Level")
plt.show()

# 3. Bar Chart
plt.figure(figsize=(6,4))
sns.countplot(x=df["Outcome"])
plt.title("Diabetes Outcome Count")
plt.xlabel("Outcome")
plt.ylabel("Number of Patients")
plt.show()

# 4. Box Plot
plt.figure(figsize=(6,4))
sns.boxplot(y=df["BMI"])
plt.title("BMI Distribution")
plt.xlabel("BMI")
plt.ylabel("BMI Value")
plt.show()

# 5. Line Chart
plt.figure(figsize=(6,4))
sns.lineplot(x=df.index[:50], y=df["Glucose"].head(50))
plt.title("Glucose Levels of First 50 Patients")
plt.xlabel("Patient Index")
plt.ylabel("Glucose Level")
plt.show()

# Preference
print("I prefer the seaborn version.")
print("Seaborn charts look cleaner, more modern, and are easier to understand visually.")
```
